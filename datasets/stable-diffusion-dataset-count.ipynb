```python
import os

def count_images(root_dir):
    count = 0
    for dirpath, _, filenames in os.walk(root_dir):
        for f in filenames:
            if f.endswith(('.jpg', '.png')):
                count += 1
    return count

REAL_DIR = os.path.join("stable-diffusion-dataset", "wiki", "wiki")
FAKE_DIR = os.path.join("stable-diffusion-dataset", "text2img", "text2img")

print("Real images:", count_images(REAL_DIR))
print("Fake images:", count_images(FAKE_DIR))
```
